```python
import torch

# 数据集
x_data = torch.tensor([[1.0],[2.0],[3.0]])
y_data = torch.tensor([[0.0],[0.0],[1.0]])
```

```python
# 模型
class logistic_regression_model(torch.nn.Module):
    def __init__(self):
        super(logistic_regression_model,self).__init__()
        self.linear = torch.nn.Linear(1,1)

    def forward(self,x):
        y_pred = torch.nn.functional.sigmoid(self.linear(x))
        return y_pred
    
model = logistic_regression_model()
```

```python
# 损失函数和优化器
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
# 训练
loss_list = []
for epoch in range(10000):
    y_pred = model(x_data)
    loss = criterion(y_pred,y_data)
    print(epoch,loss.item())
    loss_list.append(loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
import matplotlib.pyplot as plt

# 假设 loss_list 是在训练过程中记录的损失值列表

# 绘制损失变化图
plt.figure(figsize=(10, 6))
plt.plot(range(len(loss_list)), loss_list, label="Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss over Epochs")
plt.legend()
plt.grid()
plt.show()

x_test = torch.tensor([[2.5]])
print(model(x_test).item())
```
